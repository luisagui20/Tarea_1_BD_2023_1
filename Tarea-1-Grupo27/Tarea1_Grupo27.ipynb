## Crear tablas (36 pts.)

Aquí inserta el código SQL que creara las tablas de la base de datos:

### Tabla "facciones" (6 pts)

```python
%%sql
CREATE TABLE facciones (
    id INTEGER GENERATED ALWAYS AS IDENTITY,
    nombre VARCHAR(45) NOT NULL,
    descripcion VARCHAR(45) NOT NULL,
    PRIMARY KEY (id)
);
```

### Tabla "trabajos" (6 pts)

```python
%%sql
CREATE TABLE trabajos (
    id INTEGER GENERATED ALWAYS AS IDENTITY,
    nombre VARCHAR(45) NOT NULL,
    descripcion VARCHAR(45) NOT NULL,
    sueldo INTEGER NOT NULL,
    PRIMARY KEY (id)
);
```

### Tabla "personas" (6 pts)

```python
%%sql
CREATE TABLE personas
 (
    id INTEGER GENERATED ALWAYS AS IDENTITY,
    nombre VARCHAR(45) NOT NULL,
    apellidos VARCHAR(45) NOT NULL,
    fecha_nacimiento DATE NOT NULL,
    infectado BOOL NOT NULL,
    conyugue INTEGER,
    PRIMARY KEY (id),
    FOREIGN KEY (conyugue) REFERENCES personas (id)
);
```

### Tabla "cuentas_bancarias" (6 pts)

```python
%%sql
CREATE TABLE cuentas_bancarias (
    id INTEGER GENERATED ALWAYS AS IDENTITY,
    tipo_de_cuenta VARCHAR(45) NOT NULL,
    banco VARCHAR(45) NOT NULL,
    saldo INTEGER NOT NULL,
    id_persona INTEGER,
    PRIMARY KEY (id),
    FOREIGN KEY (id_persona) REFERENCES personas (id)
);
```

### Tabla "persona_pertenece_faccion" (6 pts)

```python
%%sql
CREATE TABLE persona_pertenece_faccion (
    id_persona INTEGER,
    id_faccion INTEGER,
    PRIMARY KEY (id_persona, id_faccion),
    FOREIGN KEY (id_persona) REFERENCES personas (id),
    FOREIGN KEY (id_faccion) REFERENCES facciones (id)
);

```

### Tabla "persona_tiene_trabajo" (6 pts)

```python
%%sql
CREATE TABLE persona_tiene_trabajo(
    id_persona INTEGER,
    id_trabajo INTEGER,
    estado BOOL NOT NULL,
    ultima_vez_realizado TIMESTAMP NOT NULL,
    PRIMARY KEY (id_persona, id_trabajo),
    FOREIGN KEY (id_persona) REFERENCES personas (id),
    FOREIGN KEY (id_trabajo) REFERENCES trabajos (id)
);
```

**Solución:**  Realizamos una consulta que devuelve los IDs, nombres y apellidos de todas las personas que pertenecen a una facción específica.

### Consulta 4 (6 pts)
**Nombre de la Facción y la cantidad de trabajadores activos hasta fecha X.**

Ejemplo:

| Facción     | Cantidad de trabajadores |
|----------------|--------|
| FEDRA | 1273 |
| Luciérnagas | 121 |
| Carroñeros  | 72 |
| ...  | ... |

**Solución:** Realizamos una consulta que devuelve el nombre de cada facción y la cantidad de trabajadores que pertenecen a la facción y han realizado su trabajo por última vez antes del 17 de octubre de 2019 y tienen un estado de trabajo verdadero en la tabla "persona_tiene_trabajo"

La primera línea de la consulta SELECT selecciona dos columnas diferentes: "nombre" de la tabla "facciones" y COUNT(*) que cuenta el número de registros de la tabla "persona_tiene_trabajo" que cumplen con las condiciones.

Las siguientes dos líneas de la consulta FROM especifican las tablas que se unirán: "facciones", "persona_pertenece_faccion" y "persona_tiene_trabajo".

Las líneas de la cláusula INNER JOIN especifican las condiciones para realizar la unión. La primera condición es que la columna "id_faccion" de la tabla "persona_pertenece_faccion" debe ser igual a la columna "id" de la tabla "facciones". La segunda condición es que la columna "id_persona" de la tabla "persona_pertenece_faccion" debe ser igual a la columna "id_persona" de la tabla "persona_tiene_trabajo".

La cláusula WHERE especifica dos condiciones para filtrar los registros de la tabla "persona_tiene_trabajo". La primera condición es que la columna "estado" debe ser verdadera. La segunda condición es que la fecha en la columna "ultima_vez_realizado" debe ser anterior al 17 de octubre de 2019.

La cláusula GROUP BY agrupa los resultados por el nombre de la facción.

```python
%%sql

SELECT
    facciones.nombre AS "Faccion", COUNT(*) AS "Cantidad Trabajadores"
FROM
    facciones INNER JOIN persona_pertenece_faccion ON persona_pertenece_faccion.id_faccion = facciones.id
              INNER JOIN persona_tiene_trabajo ON persona_pertenece_faccion.id_persona = persona_tiene_trabajo.id_persona 
WHERE
    persona_tiene_trabajo.estado = true
    AND date(persona_tiene_trabajo.ultima_vez_realizado) < '2019-10-17'
GROUP BY 
    facciones.nombre;


```

**Solución:** Realizamos una consulta que devuelve el nombre de cada facción y la cantidad de personas infectadas que pertenecen a cada facción, basándose en la tabla "personas" y su relación con las tablas "persona_pertenece_faccion" y "facciones".
La primera línea de la consulta SELECT selecciona dos columnas diferentes: "nombre" de la tabla "facciones" y COUNT(*) que cuenta el número de registros de la tabla "personas" que cumplen con la condición "infectado = true".

Las siguientes líneas de la consulta FROM especifican las tablas que se unirán: "personas", "persona_pertenece_faccion" y "facciones".

Las líneas de la cláusula INNER JOIN especifican las condiciones para realizar la unión. ### Consulta 6 (6 pts)
**Nombre de la Facción y su trabajo más realizado.**

Ejemplo:

|  Facción  | Trabajo | 
|--------|-------|
| FEDRA | Limpiador de cloacas |
| Luciérnagas | Médico |
| Carroñeros  | Explorador |
| ...  | ... |

**Solución:** Realizamos una consulta que muestra para cada facción, el trabajo más común entre las personas que pertenecen a esa facción.

Comenzando por la consulta externa: la cláusula SELECT inicial indica que la consulta devolverá dos columnas, una llamada "Facción" y otra llamada "Trabajo". La columna "Facción" se obtiene de la tabla "facciones" y simplemente muestra el nombre de la facción. La columna "Trabajo", por otro lado, es una subconsulta que devuelve el nombre del trabajo más común entre las personas que pertenecen a esa facción.

Dentro de la subconsulta, la cláusula SELECT también indica que se devolverá una sola columna, "nombre", de la tabla "trabajos". Luego, se realiza un INNER JOIN con la tabla "persona_tiene_trabajo" para obtener los trabajos de las personas. La condición de unión es que el ID del trabajo en la tabla "trabajos" debe ser igual al ID del trabajo en la tabla "persona_tiene_trabajo".

A continuación, se realiza otro INNER JOIN con la tabla "persona_pertenece_faccion" para obtener las personas que pertenecen a la facción. La condición de unión es que el ID de la persona en la tabla "persona_tiene_trabajo" debe ser igual al ID de la persona en la tabla "persona_pertenece_faccion" y que el ID de la facción en la tabla "facciones" debe ser igual al ID de la facción en la tabla "persona_pertenece_faccion".

Luego, se agrega una cláusula WHERE para filtrar solo los trabajos de la facción actual en la consulta externa. La condición de la cláusula WHERE es que el ID de la facción en la tabla "facciones" debe ser igual al ID de la facción en la tabla "persona_pertenece_faccion".
La subconsulta utiliza la cláusula GROUP BY para agrupar los trabajos por nombre y la cláusula ORDER BY para ordenar los trabajos por cantidad en orden descendente. La cláusula LIMIT indica que se devolverá solo el trabajo más común.
Por último, la consulta externa utiliza la cláusula ORDER BY para ordenar las filas por el nombre de la facción en orden ascendente.

```python
%%sql

SELECT facciones.nombre AS "Facción",
(SELECT trabajos.nombre
FROM 
 trabajos INNER JOIN persona_tiene_trabajo ON trabajos.id = persona_tiene_trabajo.id_trabajo
          INNER JOIN persona_pertenece_faccion ON persona_tiene_trabajo.id_persona = persona_pertenece_faccion.id_persona
WHERE facciones.id = persona_pertenece_faccion.id_faccion
GROUP BY trabajos.nombre
ORDER BY COUNT(*) DESC
LIMIT 1) AS "Trabajo"
FROM facciones
ORDER BY facciones.nombre ASC;


```

### Consulta 7 (8 pts)

Ejemplo:

|Facción | Saldo |
|--------|-|
| FEDRA | 523000 |

**Solución:** Realizamos una consulta que recupera información de varias tablas, realiza una operación de suma en una columna y muestra los resultados agrupados por una columna específica.
La consulta comienza seleccionando dos columnas: "facciones.nombre" y "SUM(cuentas_bancarias.saldo)". La primera columna, "facciones.nombre", se muestra con el alias "Faccion" y se utiliza para identificar las diferentes facciones en la tabla "facciones". La segunda columna, "SUM(cuentas_bancarias.saldo)", se muestra con el alias "Saldo" y se utiliza para sumar los saldos de todas las cuentas bancarias asociadas a cada facción.

A continuación, se unen tres tablas: "facciones", "persona_pertenece_faccion" y "cuentas_bancarias". La unión se realiza utilizando la cláusula INNER JOIN y las condiciones especificadas en "ON". La tabla "facciones" está relacionada con la tabla "persona_pertenece_faccion" mediante el campo "id". La tabla "persona_pertenece_faccion" está relacionada con la tabla "cuentas_bancarias" mediante el campo "id_persona".

Finalmente, se agrupa la consulta por el nombre de la facción utilizando la cláusula GROUP BY. Esto permite mostrar el saldo total de cada facción, en lugar de mostrar el saldo individual de cada cuenta bancaria.


```python
%%sql

SELECT facciones.nombre AS "Faccion", SUM(cuentas_bancarias.saldo) AS "Saldo"
FROM 
facciones INNER JOIN persona_pertenece_faccion ON facciones.id = persona_pertenece_faccion.id_faccion
          INNER JOIN cuentas_bancarias ON persona_pertenece_faccion.id_persona = cuentas_bancarias.id_persona
GROUP BY facciones.nombre;


```

**Nombre del Banco, cantidad de usuarios pertenecientes al banco y saldo total del banco.**

**Solución:** Realizamos una consulta que recupera información de dos tablas, se realizan dos operaciones (contar y sumar) y muestra los resultados agrupados por una columna específica (el nombre del banco).

La consulta comienza seleccionando tres columnas: "cuentas_bancarias.banco", "COUNT()" y "SUM(cuentas_bancarias.saldo)". La primera columna, "cuentas_bancarias.banco", se muestra con el alias "Banco" y se utiliza para identificar los diferentes bancos en la tabla "cuentas_bancarias". La segunda columna, "COUNT()", se muestra con el alias "Cantidad" y se utiliza para contar la cantidad de cuentas bancarias asociadas a cada banco. La tercera columna, "SUM(cuentas_bancarias.saldo)", se muestra con el alias "Saldo" y se utiliza para sumar los saldos de todas las cuentas bancarias asociadas a cada banco.

A continuación, se unen dos tablas: "cuentas_bancarias" y "personas". La unión se realiza utilizando la cláusula INNER JOIN y la condición especificada en "ON". La tabla "cuentas_bancarias" está relacionada con la tabla "personas" mediante el campo "id_persona".

Finalmente, se agrupa la consulta por el nombre del banco utilizando la cláusula GROUP BY. Esto permite mostrar la cantidad de cuentas y el saldo total asociado al banco.

```python
%%sql

SELECT 
    cuentas_bancarias.banco AS "Banco", COUNT(*) AS "Cantidad", SUM(cuentas_bancarias.saldo) AS "Saldo" 
FROM 
    cuentas_bancarias INNER JOIN personas ON cuentas_bancarias.id_persona = personas.id
GROUP BY cuentas_bancarias.banco;
```

### Consulta 9 (8 pts)
**Nombre de la Facción y el Banco más utilizado por sus habitantes**(Si hay empate, que se muestre solo uno.)

Ejemplo:
   
|Facción | Banco |
| - | - |
| FEDRA | FedraBank S.A |
| Luciérnagas | Banco Pedrito Pascal SpA |
| ... | ... |

**Solución:** Realizamos una consulta que combina información de tres tablas, utiliza una subconsulta para obtener información adicional y muestra los resultados ordenados por el nombre de la facción. La subconsulta se utiliza para obtener el nombre del banco con la mayoría de las cuentas bancarias asociadas a cada facción.

La consulta comienza seleccionando dos columnas: "facciones.nombre" y una subconsulta que utiliza la cláusula SELECT. La primera columna, "facciones.nombre", se muestra con el alias "Facción" y se utiliza para identificar las diferentes facciones en la tabla "facciones". La segunda columna, que es una subconsulta, se muestra con el alias "Banco" y se utiliza para obtener el nombre del banco con la mayoría de las cuentas bancarias asociadas a cada facción.

La subconsulta se compone de tres partes. Primero, se selecciona la columna "cuentas_bancarias.banco" de la tabla "cuentas_bancarias". A continuación, se une la tabla "persona_pertenece_faccion" con la tabla "cuentas_bancarias" mediante la cláusula INNER JOIN y la condición especificada en "ON". La tabla "persona_pertenece_faccion" está relacionada con la tabla "cuentas_bancarias" mediante el campo "id_persona". Finalmente, se utiliza la cláusula WHERE para filtrar los resultados según el "id_faccion" de la facción correspondiente.

A continuación, se agrupa la subconsulta por el nombre del banco utilizando la cláusula GROUP BY. Esto permite contar la cantidad de cuentas bancarias asociadas a cada banco. Luego, se ordena la subconsulta por la cantidad de cuentas bancarias en orden descendente utilizando la cláusula ORDER BY y se limita el resultado a una sola fila utilizando la cláusula LIMIT.

```python
%%sql

SELECT facciones.nombre AS "Facción", 
(SELECT cuentas_bancarias.banco 
FROM 
 persona_pertenece_faccion INNER JOIN cuentas_bancarias ON persona_pertenece_faccion.id_persona = cuentas_bancarias.id_persona
WHERE persona_pertenece_faccion.id_faccion = facciones.id 
GROUP BY cuentas_bancarias.banco 
ORDER BY COUNT(*) DESC 
LIMIT 1) AS "Banco"
FROM facciones
ORDER BY facciones.nombre ASC;
```

### Consulta 10 (6 pts)
**ID, nombre y segundo apellido de las personas traidoras**(Es decir, si la persona pertenece a dos o más facciones.)

|ID | Nombre | Apellido | 
|-|-------|-|
| 1 | Pedro | Pascal |
| 2 | Bella | Ramsey |
| ... | ... | ... | 

**Solución:** Realizamos una consulta que selecciona información de la tabla "personas" y utiliza una subconsulta para filtrar los resultados que tienen más de una facción asociada.

En la primera línea de la consulta, se seleccionan tres columnas de la tabla "personas": "id", "nombre" y el apellido obtenido con la función "SUBSTRING".

En la tercera línea de la consulta, se especifica una condición utilizando la cláusula WHERE. La condición se compone de una subconsulta que selecciona el "id_persona" de la tabla "persona_pertenece_faccion" y utiliza la cláusula GROUP BY para agrupar los resultados por "id_persona". La cláusula HAVING se utiliza para filtrar los resultados que tienen más de una "id_faccion" distinta asociada.

```python
%%sql

SELECT personas.id AS "ID", personas.nombre AS "Nombre", SUBSTRING(personas.apellidos FROM POSITION(' ' IN personas.apellidos)+1) AS "Apellido"
FROM personas
WHERE personas.id IN (
    SELECT persona_pertenece_faccion.id_persona
    FROM persona_pertenece_faccion
    GROUP BY id_persona
    HAVING COUNT(DISTINCT id_faccion) > 1
);
```
